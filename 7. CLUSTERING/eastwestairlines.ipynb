# Hierarchy Clustering


```python
# create dendrogram
plt.figure(figsize = (16, 10));
dendrogram = sch.dendrogram(sch.linkage(al, method='average'))
```

```python
# Agglomerative Clustering
hc = AgglomerativeClustering(n_clusters=5, affinity = 'euclidean', linkage = 'average')
# save clusters for chart
y_hc = hc.fit_predict(al)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
al1 = pd.concat([al,Clusters], axis=1)
al1
```

```python
al1.sort_values("Clusters")
```

# K-Means

```python
km = KMeans(n_clusters=14,random_state=0)
km.fit(al)
```

```python
km.inertia_
```

```python
wcss = []
for i in range(1, 19):
    km = KMeans(n_clusters=i,random_state=0)
    km.fit(al)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 19), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(7, random_state=0)
clusters_new.fit(al)
```

```python
clusters_new.labels_
```

```python
# Converting array to dataframe
df = pd.DataFrame(clusters_new.labels_, columns =['Cluster ID'])
al2 = pd.concat([al,df], axis=1)
al2
```

```python
al3=al2.drop(['ID#'], axis=1)
al3
```

```python
al3.groupby('Cluster ID').agg(['mean']).reset_index()
```

```python
al3['Cluster ID'].value_counts()
```

```python
stscaler = StandardScaler().fit(abc)
X1 = stscaler.transform(abc)
```

```python
dbscan = DBSCAN(eps=0.82, min_samples=6)
dbscan.fit(X1)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['Cluster'])
al4 = pd.concat([al,cl],axis=1)
al4
```
