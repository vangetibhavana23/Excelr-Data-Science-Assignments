```python
bs_new = bs.pivot_table(index='UserId', columns='Title', values='Rating')
bs_new
```

```python
#Impute those NaNs with 0 values
bs_new.fillna(0, inplace=True)
bs_new.index = bs['UserId'].unique()
bs_new
```

# Calculating Cosine Similarity between Users

```python
user1 = 1 - pairwise_distances( bs_new.values, metric='cosine')
user1
```

```python
np.fill_diagonal(user1, 0)
#Store the results in a dataframe
user = pd.DataFrame(user1)
user
```

```python
#Set the index and column names to user ids
user.index = bs.UserId.unique()
user.columns = bs.UserId.unique()
user
```

# Considering the most similar customer

```python
def recommend(custID):
    simID = user.loc[custID].idxmax()
    simID_bs = bs[bs['UserId'] == simID].Title
    custID_bs = bs[bs['UserId'] == custID].Title
    return set(simID_bs) - set(custID_bs)
```

```python
recommend(276813)
```

```python
recommend(507)
```

```python
recommend(3462)
```
