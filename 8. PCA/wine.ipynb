```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster
```

```python
w = pd.read_csv("wine.csv")
w
```

```python
w['Type'].value_counts()
```

```python
ws = w.drop(['Type'],axis=1)
ws.head(5)
```

```python
ws.info()
```

```python
ws.isna().sum()
```

```python
ws.describe()
```

## Principal Component Analysis

```python
# Converting data to numpy array
ws_ary=ws.values
ws_ary
```

```python
# Normalizing the numerical data
ws1=scale(ws_ary)
ws1
```

```python
pca = PCA(n_components = 13)
pca_values = pca.fit_transform(ws1)
pca_values
```

```python
var = pca.explained_variance_ratio_
var
```

```python

var1 = np.cumsum(np.round(var,decimals = 8)*100)
print("The amount of variance that each PCA explains is ",var1)
```

```python
pca.components_
```

```python
plt.plot(var1,color="red")
```

```python
ws_new = pd.concat([pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3']), w['Type']], axis = 1)
ws_new
```

```python
# Visualization of PCAs
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data = ws_new)
```

Checking with other Clustering Algorithms
# 1. Heirarchial Clustering

```python

# create dendrogram
plt.figure(figsize = (12, 7));
dendrogram = sch.dendrogram(sch.linkage(ws1, method='complete'))
```

```python
#plt.figure(figsize = (13,8));
#dendrogram = sch.dendrogram(sch.linkage(ws, method='ward'))
# create clusters
hc1 = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'average')
# save clusters for chart
y2 = hc1.fit_predict(ws)
cc = pd.DataFrame(y2,columns=['Clusters'])
cc.head(6)
```

```python
ws2 = pd.concat([w,cc],axis=1)
ws2.head(7)
```

```python

ws2['Clusters'].value_counts()
```

# 2. K-Mean Clustering

```python
wcss = []
for i in range(1, 8):
    kmeans = KMeans(n_clusters=i,random_state=2)
    kmeans.fit(ws)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 8), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm
cc2 = KMeans(3, random_state=27)
cc2.fit(ws)
```

```python
cc2.labels_
```

```python
# Converting array to dataframe
df2 = pd.DataFrame(cc2.labels_, columns =['clusters'])
ws3 = pd.concat([ws,df2], axis=1)
ws3
```

```python
ws3.groupby('clusters').agg(['mean'])
```
