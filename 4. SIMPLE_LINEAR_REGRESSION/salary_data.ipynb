```python
abc = abc.rename({'YearsExperience':'Experience'},axis=1)
abc                          #Rename function
```

```python
sns.regplot(x=abc['Experience'],y=abc['Salary'])
```

```python
model1 = smf.ols("Salary~Experience",data = abc).fit()
print(model1.params,'\n''\n',model1.tvalues,'\n''\n', model1.pvalues)
```

```python
#R squared values
(model1.rsquared,model1.rsquared_adj)
```

```python
model1.summary()
```

```python
#Prediction for Salary hike
newdata=pd.Series([7,15])
d_p=pd.DataFrame(newdata,columns=['Experience'])
d_p
```

```python
model1.predict(d_p)
```

```python
model2 = smf.ols("Salary~np.log(Experience)",data = abc).fit()   #using log
print(model2.params,'\n''\n',model2.tvalues,'\n''\n', model2.pvalues)
```

```python
#R squared values
(model2.rsquared,model2.rsquared_adj)
```

```python
model2.summary()
```

```python
model3 = smf.ols("Salary~np.square(Experience)",data = abc).fit()   #using square
print(model3.params,'\n''\n',model3.tvalues,'\n''\n', model3.pvalues)
```

```python
#R squared values
(model3.rsquared,model3.rsquared_adj)
```

```python
model3.summary()
```
